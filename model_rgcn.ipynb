## Observation of dataset
In order to have better understanding of our dataset. First, we check the max amount of the dialogue act annotation types and max amount of relations that one utterance can have.

```python
def extract_relation_types(dir_path,txt_files):
    relation_types = set() 
    for file_name in txt_files:
        txt_path = dir_path + file_name
        with open(txt_path, 'r') as file:
            for line in file:
                parts = line.strip().split()
                if len(parts) == 3:
                    _, relation_type, _ = parts
                    relation_types.add(relation_type)

    return relation_types

def count_max_relations_per_index(dir_path, txt_files):
    max_relations_per_file = {}

    for file_name in txt_files:
        txt_path = dir_path + file_name
        index_relation_count = {} 

        with open(txt_path, 'r') as file:
            for line in file:
                parts = line.strip().split()
                if len(parts) == 3:
                    source_index = int(parts[0]) 
                    index_relation_count[source_index] = index_relation_count.get(source_index, 0) + 1

        max_relations = max(index_relation_count.values(), default=0)
        max_relations_per_file[file_name] = max_relations

    return max_relations_per_file
```

## Feature selection
In our model, we apply KeyBERT for feature selections, after keywords extracted, we add information of discourse graph to contruct our training and test dataset..

```python
import json
import csv
from keybert import KeyBERT

def read_discourse_graph(txt_file_path, offset):
    """
    Reads a discourse graph from a .txt file and returns a dictionary of edges,
    adjusting indices based on the provided offset.
    """
    edges = {}
    with open(txt_file_path, 'r', encoding='utf-8') as file:
        for line in file:
            if line.strip():
                source, relation, target = line.split()
                source, target = int(source) + offset, int(target) + offset
                if source not in edges:
                    edges[source] = []
                edges[source].append((target, relation))
    return edges

def clean_text(text):
        return text.replace("<vocalsound>", "")

def extract_features(json_files, csv_file, dir_path):
    kw_model = KeyBERT()
    offset = 0  # Initialize the offset
    exclude_words = {'um','Um','vocalsound','mmm','hmm','uh','oh','uh um','hm'} # add some filtered words

    with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)

        for file_name in json_files:
            base_name = file_name.split('.')[0]
            json_path = dir_path + file_name
            txt_path = dir_path + base_name + '.txt'

            # Read discourse graph with the current offset
            edges = read_discourse_graph(txt_path, offset)

            with open(json_path, 'r', encoding='utf-8') as file:
                data = json.load(file)

            writer.writerow([base_name])  # Write file name as a label

            for entry in data:
                index = entry['index'] + offset
                text = clean_text(entry['text'])
                keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 5), stop_words='english', top_n=10, use_maxsum=True, nr_candidates=20)
                keywords = [kw[0] for kw in keywords if kw[0].lower() not in exclude_words]

                # Add discourse relations with updated indices
                related_indices = edges.get(index, [])
                relations = [f"{target}-{relation}" for target, relation in related_indices]

                # Combine all features into one row
                # row = [index] + keywords + relations
                row = relations + keywords
                writer.writerow(row)

            # Update offset for the next file
            offset += len(data)
```

```python
import json
import pandas as pd
import re

# Load JSON labels
with open('/Users/aphrodite/Downloads/intro-to-ml-main/training_labels.json', 'r') as file: ## change!!!
    labels_data = json.load(file)

# Read CSV file as text
with open('/Users/aphrodite/Downloads/intro-to-ml-main/output_train_keyword_test5.csv', 'r') as file:  ## change!!!
    lines = file.readlines()

# Initialize lists for texts and relations
labels = []
texts = []
relations = [[], [], [], [], [], []]  # Lists for relation_1, relation_2, relation_3, relation_4, 5, 6
current_labels = []

for line in lines:
    line = line.strip()
    parts = line.split(',')

    # Initialize variables to store relation and text data for each line
    line_relations = []
    line_text = ""

    is_new_section = parts[0] in labels_data

    # Check and extract relations and text from each part
    for part in parts:
        if bool(re.match(r'^\d+-', part)) and not is_new_section:
            line_relations.append(part)
        elif not is_new_section:
            line_text += part + " "  # Combine text parts, add space for separation

    # Append extracted relations to the relations list, filling in blanks as needed
    if not is_new_section:
        for i in range(6):
            relations[i].append(line_relations[i] if i < len(line_relations) else "")

    # Handle new section for labels
    if is_new_section:
        if current_labels:
            labels.extend(current_labels[:len(texts)])
            texts.extend([""] * (len(current_labels) - len(texts)))
            for rel_list in relations:
                rel_list.extend([""] * (len(current_labels) - len(rel_list)))
        current_labels = labels_data[parts[0]]
    else:
        # Append the extracted text
        texts.append(line_text.strip())  # Strip to remove trailing spaces

# Add the last section
labels.extend(current_labels[:len(texts)])
texts.extend([""] * (len(current_labels) - len(texts)))
for rel_list in relations:
    rel_list.extend([""] * (len(texts) - len(rel_list)))

# Check lengths of lists
print(len(texts), len(relations[0]), len(relations[1]), len(relations[2]), len(relations[3]), len(labels))

# Create DataFrame
df = pd.DataFrame({
    'text': texts,
    'relation_1': relations[0],
    'relation_2': relations[1],
    'relation_3': relations[2],
    'relation_4': relations[3],
    'relation_5': relations[4],
    'relation_6': relations[5],
    'label': labels
})

df
```

Now we train RGCN model.

```python
import torch
from sklearn.feature_extraction.text import TfidfVectorizer
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import RGCNConv
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

vectorizer = TfidfVectorizer(max_features=100)
x = vectorizer.fit_transform(df['text']).toarray()  # edge feature

edge_index = []
edge_type = []
num_relations = 6

# Construct edges
for idx, row in df.iterrows():
    for rel_idx in range(num_relations):
        relation = row[f'relation_{rel_idx+1}']
        if relation:
            target_index, _ = relation.split('-', 1)
            edge_index.append([idx, int(target_index)])
            edge_type.append(rel_idx)

edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
edge_type = torch.tensor(edge_type, dtype=torch.long)

# Labels
y = torch.tensor(df['label'].values, dtype=torch.long)


class RGCN(torch.nn.Module):
    def __init__(self, in_channels, out_channels, num_relations):
        super(RGCN, self).__init__()
        self.conv1 = RGCNConv(in_channels, 32, num_relations)
        self.conv2 = RGCNConv(32, 64, num_relations)
        self.conv3 = RGCNConv(64, out_channels, num_relations)
        self.dropout = torch.nn.Dropout(0.5)

    def forward(self, data):
        x, edge_index, edge_type = data.x, data.edge_index, data.edge_type

        x = F.relu(self.conv1(x, edge_index, edge_type))
        x = self.dropout(x)
        x = F.relu(self.conv2(x, edge_index, edge_type))
        x = self.dropout(x)
        x = self.conv3(x, edge_index, edge_type)

        return F.log_softmax(x, dim=1)

# Initialize model
model = RGCN(100, len(torch.unique(y)), num_relations)

# Split dataset
train_idx, test_idx = train_test_split(range(len(df)), test_size=0.5, random_state=42)
train_mask = torch.zeros(len(df), dtype=torch.bool)
train_mask[train_idx] = True
test_mask = torch.zeros(len(df), dtype=torch.bool)
test_mask[test_idx] = True

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = RGCN(100, len(torch.unique(y)), num_relations).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
data = Data(x=torch.tensor(x, dtype=torch.float).to(device),
            edge_index=edge_index.to(device),
            edge_type=edge_type.to(device),
            y=y.to(device),
            train_mask=train_mask, 
            test_mask=test_mask)

# Prepare weights
weight_for_class_0 = 1.0
weight_for_class_1 = 3.0  # add the weight of label 1

#  Convert to Tensor and move to the correct device
class_weights = torch.tensor([weight_for_class_0, weight_for_class_1], dtype=torch.float32).to(device)

# # Use weighted loss function
criterion = torch.nn.CrossEntropyLoss(weight=class_weights)
```

```python
model.train()
for epoch in range(200):  # epoch
    optimizer.zero_grad()
    out = model(data)
    # loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])  
    loss = criterion(out[data.train_mask], data.y[data.train_mask])  
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```

```python
model.eval()
_, pred = model(data).max(dim=1)
correct = float(pred[data.test_mask].eq(data.y[data.test_mask]).sum().item())
accuracy = correct / data.test_mask.sum().item()
y_test = data.y[data.test_mask].numpy()
y_pred = pred[data.test_mask].numpy()
# print(classification_report(y_test, y_pred))
# print(f'Accuracy: {accuracy}')
```
